```python
df_branches = pd.read_csv(data_dir / "t_branches.csv").drop("Unnamed: 0", axis="columns")
df_cities = pd.read_csv(data_dir / "t_cities.csv").drop("Unnamed: 0", axis="columns")
df_products = pd.read_csv(data_dir / "t_products.csv").drop("Unnamed: 0", axis="columns")
df_sales = pd.read_csv(data_dir / "t_sales.csv").drop("Unnamed: 0", axis="columns")
```

```python
df_names_sales = df_sales.merge(right=df_products, left_on="Номенклатура", right_on="Ссылка").drop("Ссылка", axis="columns")
df_names_sales.head(3)
```
