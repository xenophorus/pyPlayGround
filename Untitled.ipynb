```python
df_names_sales = df_sales.merge(right=df_products, left_on="Номенклатура", right_on="Ссылка").drop("Ссылка", axis="columns")
df_names_sales.head(3)
```

```python
a = df_names_sales['Период']
# a.apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
df_names_sales['Период'] = pd.to_datetime(a, yearfirst=True)
```

```python
df_names_sales.columns = ["date_time", "branch", "product_uuid", "quantity", "sale", "product_name"]
```

```python
df_names_sales["dates"] = df_names_sales["date_time"].dt.date
```

```python
df_names_sales = df_names_sales[df_names_sales['quantity'] > 0]
```

```python
df_names_sales['price_by_unit'] = df_names_sales['sale'] / df_names_sales['quantity']
```
